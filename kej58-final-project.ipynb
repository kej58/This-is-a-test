```python
#import pandas
import pandas as pd
#import matplotlib for graphing
import matplotlib.pyplot as plt

#We first import the data from a csv file and display the first 10 rows
blotter_data = pd.read_csv('BlotterData.csv')
blotter_data.head()
```

```python
#count incident that happened in each location in numerical order
blotter_data['INCIDENTLOCATION'].value_counts().sort_values(ascending=True)
```

```python
#Use the matplotlib for graphing
plt.rcParams['figure.figsize']=(20,12)
plt.title('Incident Reported', fontweight=20,fontsize=20)
plt.style.use("seaborn")
#Find out the most reported crime
rate=blotter_data.INCIDENTHIERARCHYDESC.value_counts()
#plot it as a bar graph
rate.iloc[:10].sort_values().plot.pie()
plt.xticks(rotation=90)
plt.show()
```

```python
#report the first 10 neigborhood that report 99
report = blotter_data[blotter_data.INCIDENTHIERARCHYDESC.str.contains('99')];
report.INCIDENTNEIGHBORHOOD.value_counts()
report.iloc[:10]
```

```python
#report the first 10 neigborhood that report theft
report = blotter_data[blotter_data.INCIDENTHIERARCHYDESC.str.contains('THEFT')];
report.INCIDENTNEIGHBORHOOD.value_counts()
report.iloc[:10]
```

```python
#report the first 10 neigborhood that report simple assault
report = blotter_data[blotter_data.INCIDENTHIERARCHYDESC.str.contains('SIMPLE ASSAULT')];
report.INCIDENTNEIGHBORHOOD.value_counts()
report.iloc[:10]
```

```python
#count the first 20 neighborhood with the least crime reported
less_crime=blotter_data.groupby('INCIDENTNEIGHBORHOOD').size().sort_values(ascending=True)
less_crime.head(20)
```

```python
#plot the region with the least incident reported
data = less_crime.head(20).plot(figsize=(20,10),kind='barh',title="Neighborhood with least incident")
plt.show()
```
